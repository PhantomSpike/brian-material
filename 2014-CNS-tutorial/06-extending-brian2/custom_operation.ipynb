```python
from brian2 import *
```

```python
tau = 10*ms
G = NeuronGroup(10, '''dv/dt = (-v + I)/tau + tau**-0.5*xi : 1
                       I : 1 (shared)''')
G.freq = 100*Hz

updater = G.custom_operation('I = 10*rand()',  #draw a new random value for the input
                             when=Clock(dt=100*ms))

```

```python
mon = StateMonitor(G, 'v', record=True)
run(1*second)
```

```python
_ = plot(mon.t/ms, mon.v.T, color='gray')
```
